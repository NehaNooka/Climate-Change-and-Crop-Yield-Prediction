```python
# 1. Summary Statistics
summary_stats = df.describe()
print(summary_stats)
```

```python
# 2. Data Cleaning
# Check for missing values
df.isnull().sum()
```

```python
# 3. Univariate Analysis
# Plot histograms for numerical variables
numerical_cols = ['Year', 'Crop_Production_Value', 'Emission_Total_Value', 'Production_Value_$',
                  'Nutrients_Value', 'TMP_Value', 'Area Value']
df[numerical_cols].hist(bins=20, figsize=(15, 10))
plt.show()
```

```python
# 4. Bivariate Analysis
# Plot bar charts for categorical variables
categorical_cols = ['Area', 'Prod_type', 'Crop_Name', 'Crop_Production_Flag_Description',
                    'Emission_Total_Flag_Description', 'Production_Value_Type', 'Prod_Output_Flag_Description',
                    'Nutrients_type', 'TMP_Description', 'Area_type', 'Land_Use_Flag_Description']
for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=col, color="#9b59b6")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Scatter plot between numerical variables
sns.pairplot(df[numerical_cols])
plt.show()
```

```python
# 5. Temporal Analysis
# Line plot for variables over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Crop_Production_Value')
plt.xlabel('Year')
plt.ylabel('Crop Production Value')
plt.title('Crop Production Value over Time')
plt.show()
```

```python
# 6. Outlier Detection
# Box plot for numerical variables
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[numerical_cols])
plt.show()
```

```python
# 7. Correlation Analysis
# Correlation matrix
correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# 8. Multivariate Analysis
# Pairwise scatter plot with hue for categorical variable
plt.figure(figsize=(12, 8))
sns.pairplot(df, hue='Prod_type')
plt.show()
```

```python
# 9. Distribution Comparison
# Kernel density estimation plot
plt.figure(figsize=(12, 6))
for col in numerical_cols:
    sns.kdeplot(data=df, x=col, label=col, shade=True)
plt.legend()
plt.xlabel('Value')
plt.title('Kernel Density Estimation Plot')
plt.show()
```
